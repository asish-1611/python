# Types of Features

---

## 🔹 1. Numerical Features (also called Quantitative Features)
Features that represent **numbers**.

**Examples:** Age, Salary, Temperature, Height

**Types of numerical features:**
- **Continuous:** Can take any value in a range (e.g., weight = 65.23 kg, 70.5 kg, etc.)
- **Discrete:** Only specific values (e.g., number of children = 1, 2, 3)

---

## 🔹 2. Categorical Features (Qualitative Features)
Features that represent **categories** or **labels**.

**Examples:** Gender (Male/Female), City (Paris, Tokyo)

**Types of categorical features:**
- **Nominal:** No natural order (e.g., Color: Red, Blue, Green)
- **Ordinal:** Natural order exists (e.g., Education level: High School < Bachelor < Master)

---

## 🔹 3. Binary Features
Special case of **categorical**: only **two values**.

**Examples:** Yes/No, 0/1, True/False

---

## 🔹 4. Text Features
When a feature contains **free-form text**.

**Examples:** Product reviews, Tweets, Messages

*(These usually require special processing like tokenization.)*

---

## 🔹 5. Datetime Features
Features that represent **dates and times**.

**Examples:** Date of Birth, Transaction Timestamp

*(We often extract "day", "month", "weekday", etc. from these.)*

---

## 🔹 6. Image/Audio Features (in advanced tasks)
When the input is an **image** or a **sound clip**.

*(Normally used in deep learning fields like computer vision or speech recognition.)*

```python
df2
```

```python
df2.dropna(axis=0)
```

```python
pd.isna(df2['one'])
#this returns true if there is a missing value(ie.Nan) else false
```

## CSV

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```python
type(data)
# StringIO(data) simulates a file from the data string.
# StringIO creates an in-memory file-like object from a string.
# If you have CSV data in the form of a string (e.g., coming from an API, or user input).
# If you're testing or working with small datasets that are not stored in actual files.
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns
df= pd.read_csv(StringIO(data),usecols=['col1','col2'])
# df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
df
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
# The escape character ('\\') is used to escape special characters that would otherwise have a special meaning in CSV files (e.g., quotes).
# Here, the escape character \\ tells pandas that the quote characters around Bob are part of the data and not delimiters or quotes that define the boundaries of the string.
```
