## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[2,10,4,8,10]
y1=[4,6,8,9,10]
plt.subplot(1,2,1)
plt.plot(x,y,color="r",linestyle="--",marker="D",markersize=10)
plt.subplot(1,2,2)
plt.plot(x,y1,color="g",linestyle="-.",marker="+")
plt.title("Line-Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
# print(x)
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
marks ={"asish" : 76 , "burra" :67 ,"shaik" :89 ,"amulya" : 47}
x1=np.sort(list(marks.keys()))
y1=np.sort(list(marks.values()))

colours =['red','green','blue', 'yellow']
plt.bar(x1,y1,color=colours)
plt.show()
```

```python
#for horizontal bar plot 
labels='marks'
plt.barh(x1,y1,color=colours,label=labels)
plt.xlabel("Marks")
plt.ylabel("students")
plt.title("K23sh")

plt.legend()
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]
 
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.legend(["round1","Round2"])
plt.show()
```

```python
import pandas as pd
df=pd.DataFrame([['A', 10, 20, 10, 26], ['B', 20, 25, 15, 21], ['C', 12, 15, 19, 6],
                   ['D', 10, 18, 11, 19]], columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])
df
df.plot(x='Team',kind='bar',stacked=True,title='Stacked Bar Graph')
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
A = np.array([3,6,9,4,2,5])
X = np.arange(6)
 
# plot the bars
plt.bar(X, A, color = 'r')
plt.bar(X, -A, color = 'b')
plt.title("Back-to-Back Bar Chart")
plt.grid(True)
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
cars=["BMW","Mercedes","Honda","Tesla"]
data=[10,20,40,50]
plt.pie(data,labels=cars,colors=["yellow","Pink","magenta","Green"])
plt.show()
```

```python
cars=["BMW","Mercedes","Honda","Tesla"]
data=[10,20,40,50]
plt.pie(data,labels=cars,colors=["yellow","Pink","magenta","Green"],radius=1,autopct='%0.2f%%')
plt.pie([1],colors=['w'],radius=0.5)
plt.show()
```
