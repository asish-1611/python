### Correlation with Heatmap

 A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types.
 It is very important in Feature Selection

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

## JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
### Univariate Analysis



```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature


```python
sns.histplot(df['tip'],kde=True)
```

```python
sns.histplot(df['tip'],kde=False,bins=10)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [5,4,9,6,7]})
sns.lineplot(x='x', y='y', data=data)
plt.title("Seaborn Line Plot")
plt.show()
```

```python
import pandas as pd
data = pd.DataFrame({'category': ['A', 'B', 'C', 'D', 'E'], 'values': [3, 7, 2, 5, 4]})
sns.barplot(x='category', y='values', data=data)
plt.title("Seaborn Bar Plot")
plt.show()
```

```python
import numpy as np
data = np.random.randn(1000)
sns.histplot(data, bins=30,kde=True)#kernel Density Estimation.
plt.title("Seaborn Histogram with KDE")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
data={
    "column":["A","B","C","D"],
    "Values":[10,20,30,40]
}
r=pd.DataFrame(data)
r
sns.lineplot(x="column",y="Values",data=r,marker="o")
sns.barplot(x="column",y="Values",data=r,
            hue="column")

```

Seaborn also helps us in doing the analysis on Categorical Data points. 

- boxplot
- violinplot

```python
## Count plot

sns.countplot(x='sex', data=df)
```

```python
## Count plot

sns.countplot(y='sex',data=df)
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='smoker', y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow',hue="day")
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow',hue="day")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Plotting strip plot with 'species' on the x-axis and 'sepal_length' on the y-axis
ax = sns.stripplot(x='species', y='sepal_length', data=iris, palette='Set2',hue='species',legend=True)

# Optional: Add labels and a title for better presentation
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Strip Plot of Sepal Length by Species')

plt.show()
```
