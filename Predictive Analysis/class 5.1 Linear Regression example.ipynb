Q. Predict marks based on no of hours studied

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Hours_studied' : [30 ,28 ,20 ,12 ,33 ,35 ,40 ,45 ,50 ,55],
    'Marks' : [45, 37, 25, 20, 50, 53, 76, 88, 90, 98],
}
df = pd.DataFrame(data)

y = df['Marks'] #dependent
x = df[['Hours_studied']] #independent
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print("hours",x_test)
print("marks\n",y_test)
print("Actual ")
print("predicted hours:",y_pred)
```

```python
plt.scatter(x_train,y_train,color='blue',label='Training data')
plt.scatter(x_test,y_test,color='green',label='Test data')

plt.plot(x,model.predict(x),color='red',linewidth=1, label='Regression line')
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title('Marks vs hours studied')
plt.legend()
plt.grid(True)
plt.show()
plt.tight_layout()
```
